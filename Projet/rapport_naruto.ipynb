Les premiers jeux de ce genre, comme *Street Fighter II* (1991) ou *Mortal Kombat* (1992), ont posé les bases du gameplay compétitif en 2D . Deux personnages s’affrontent sur un même plan, avec des coups, des combos et des animations spectaculaires. Ces jeux étaient souvent limités par les contraintes techniques de l’époque (graphismes pixelisés, animations simples), mais ils ont su créer une véritable culture du duel et de la stratégie.

Avec le temps, les jeux de combat 2D ont évolué, intégrant des sprites plus détaillés, des effets visuels plus riches et des systèmes d’animation de plus en plus sophistiqués. Aujourd’hui, même si la 3D domine le marché, le style 2D reste très apprécié pour sa lisibilité, sa nervosité et son côté rétro. Mon projet s’inscrit dans cette tradition : il reprend les principes fondamentaux du combat 1 vs 1, tout en y ajoutant une touche personnelle inspirée de l’univers de Naruto.

### **3.2. La Solution : OpenCV**
<a id="3.2"></a>

Face à ce problème, j’ai décidé d’automatiser le nettoyage des sprites grâce à la bibliothèque **OpenCV** (`cv2`). Plutôt que de supprimer le bleu à la main sur plus de 260 images (ce qui aurait été interminable et source d’erreurs), j’ai écrit un script Python qui s’exécute au démarrage du jeu.

Ce script lit chaque image de sprite, détecte précisément la couleur du fond (une teinte bleue comprise entre 100 et 130), puis remplace tous les pixels de cette couleur par des pixels transparents. Ce procédé m’a permis de gagner un temps considérable et d’assurer une cohérence parfaite sur l’ensemble des sprites.

Mais tout n’a pas été simple : par exemple, le Rasengan (l’attaque spéciale de Naruto) est bleu cyan, très proche du bleu du fond. Au début, mon script effaçait aussi le Rasengan par erreur ! J’ai donc dû affiner les seuils de détection pour préserver les effets spéciaux tout en supprimant uniquement le fond.

Grâce à cette méthode, je peux désormais intégrer n’importe quelle nouvelle planche de sprites trouvée sur internet : le jeu la nettoie automatiquement, ce qui rend le projet évolutif et facilement personnalisable. Ce travail sur les assets m’a permis de mieux comprendre l’importance de l’automatisation et de la robustesse dans la gestion des ressources graphiques pour un jeu vidéo.

## **4.  Programmation Orientée Objet, Machine à États et Hitbox**
<a id="4"></a>


### **4.1. Logique POO et Principales Classes**
<a id="4.1"></a>

Le projet repose sur une architecture orientée objet afin de séparer clairement les responsabilités et de faciliter l’évolution du code. Chaque élément du gameplay est encapsulé dans une classe dédiée, ce qui améliore la lisibilité, la maintenance et l’extensibilité du système.

**Personnage** : Représente un combattant et centralise les données persistantes (position, points de vie, vitesse, sprite). Il délègue une partie de son comportement à d’autres composants, notamment l’état courant et les hitboxes actives.

**Hitbox** : Modélise une zone de collision, associée soit à un personnage (hurtbox), soit à une attaque (hitbox offensive). Elle est utilisée uniquement pour la détection des interactions physiques, sans logique de gameplay directe, ce qui permet de réutiliser le système pour les projectiles ou attaques spéciales.

**État** : Implémente une machine à états contrôlant le comportement du personnage (idle, courir, attaquer, sauter, etc.). Chaque état définit ses propres règles de mise à jour, d’entrées utilisateur et de transitions, ce qui évite les conditions complexes dans la classe Personnage.

**Jeu** : Joue le rôle de contrôleur principal. Il gère la boucle de jeu, relaie les entrées utilisateur, met à jour les objets et orchestre les interactions entre les différents composants.

### **4.2. Machine à États**
<a id="4.2"></a>

Chaque personnage est piloté par une machine à états finie (FSM) regroupant les états essentiels (idle, run, attack, jump, hurt, etc.). Chaque état encapsule le comportement, l’animation et les règles de mise à jour ; les transitions sont déclenchées par des événements (entrée utilisateur, collision, fin d’animation) et soumises à conditions/priorités pour garantir cohérence et fluidité. ### **4.3. Système de Hitbox**
<a id="4.3"></a>

Le système repose sur deux concepts simples :

La **hurtbox**, qui représente la zone où un personnage peut être touché, et la **hitbox**, qui correspond à la zone active d’une attaque.Les boîtes sont définies à l’aide d’offsets relatifs au sprite et sont recalculées à chaque frame via la fonction get_hitbox().

Seules les hitboxes marquées comme actives pendant les frames d’impact peuvent appliquer des effets (dégâts, recul, passage à l’état « hurt »), après un test de collision.

Cette approche présente plusieurs avantages : elle est simple à déboguer (grâce à l’affichage temporaire des boîtes) et facilement extensible, notamment pour gérer des projectiles ou des attaques spéciales.


Exemple de get_hitbox (méthode de la classe Fighter) :

```python
def get_hitbox(self):
    frames = self.get_current_frames()
    frame = frames[self.frame_idx % len(frames)]
    w, h = frame.get_size()
    hb_w = int(w * 0.6)
    hb_h = int(h * 0.8)
    hb_x = int(self.x + (w - hb_w) // 2 + getattr(self, 'hitbox_offset_x', 0))
    hb_y = int(self.y + (h - hb_h) // 2 + getattr(self, 'hitbox_offset_y', 0))
    return pygame.Rect(hb_x, hb_y, hb_w, hb_h)
```

Le système sépare la hurtbox (zone vulnérable) et la hitbox (zone active d’attaque). get_hitbox() calcule chaque frame une hitbox proportionnelle à la frame, l’ajuste avec des offsets et renvoie un pygame.Rect. On teste ensuite l’intersection (p.ex. pygame.Rect.colliderect) pour appliquer dégâts, recul et passer en état « hurt ».


## 5.2 Système de garde<img src="assets/naruto_frames/sprite_043.png" alt="Exemple de Sprite Sheet avec fond bleu" width="40">
<a id="5.2"></a>

Le système de garde est un élément primordial dans les jeux de combat, car il structure la dimension défensive et renforce la profondeur stratégique des affrontements.

La garde permet au joueur de bloquer les attaques adverses lorsqu’elle est activée au bon moment. Elle repose sur une barre de défense qui diminue à chaque impact bloqué : tant que cette barre est supérieure à zéro, les dégâts sont réduits ou annulés. Lorsqu’elle est épuisée, la garde devient temporairement inefficace et les attaques passent normalement.

Techniquement, le blocage s’appuie sur le système de hitbox existant : lors d’une collision entre une attaque et le personnage en état de garde, l’impact est traité comme bloqué et consomme de la ressource, laquelle se régénère progressivement hors garde. Pour ce faire on vas utiliser la fonction **get_hitbox()** pour réaliser ceci.

```python
# Gestion de la hitbox pour la détection de collision
 def get_hitbox(self):
     frames = self.get_current_frames()
     if self.frame_idx >= len(frames):
         self.frame_idx = 0
     frame = frames[self.frame_idx]
     w, h = frame.get_size()
     hitbox_w = int(w * 0.6)
     hitbox_h = int(h * 0.8)
     hitbox_x = self.x + (w - hitbox_w) // 2
     hitbox_y = self.y + (h - hitbox_h) // 2
     return pygame.Rect(hitbox_x, hitbox_y, hitbox_w, hitbox_h)
```
La fonction get_hitbox() calcule dynamiquement la zone de collision du personnage à partir du sprite de l’animation en cours. Elle ajuste la taille de la hitbox (plus petite que le sprite) et la centre sur la position du personnage, afin d’obtenir une détection de collisions plus précise et cohérente avec l’animation.
